```python
def Powerball():
    """Powerball Function is a program designed similar to that of the Powerball Lottery."""
    
    # Displaying the general rules of the powerball lottery so the user can have an understanding of how to play
    print("**************************************** Rules **************************************** \n\
    1) Each ticket costs $2.00 to play \n\
    2) For each ticket, you will select 5 non-powerball numbers that are between 1 and 69. \n\
         **Note: For the 5 non-powerball numbers, there cannot be any duplicating numbers on the same ticket. \n\
    3) Then you will select a powerball number that is between 1 and 26. \n\
    4) Winnings are based on the number of matching numbers that you have with the winnings numbers \n\n\
    Possible winning matching combinations ................................. Corresponding Winnings \n\
        5 out of 5 non-Powerball numbers AND Powerball number .............. Jackpot ($100,000,000.00) \n\
        5 out of 5 non-Powerball numbers WITHOUT Powerball number .......... $1,000,000.00 \n\
        4 out of 5 non-Powerball numbers AND Powerball number .............. $50,000.00 \n\
        4 out of 5 non-Powerball numbers WITHOUT Powerball number .......... $100.00 \n\
        3 out of 5 non-Powerball numbers AND Powerball number .............. $100.00 \n\
        3 out of 5 non-Powerball numbers WITHOUT Powerball number .......... $7.00 \n\
        2 out of 5 non-Powerball numbers AND Powerball number .............. $7.00 \n\
        1 out of 5 non-Powerball numbers AND Powerball number .............. $4.00 \n\
        Powerball number only .............................................. $4.00 \n\
*****************************************************************************************")

    # importing the random module for the program to randomly generate the winning lottery numbers
    # initializing winnings to 0, so that a running total of the overall winnings can be kept
    # initializing tickets bought to 0, so that a running total of the number of tickets bought can be kept
    # creating an empty dictionary to store the ticket number and its corresponding list of user selected lottery numbers
    import random
    winnings = 0
    tickets_bought = 0
    User_tickets = dict()
    
    
    while True:
        # While loop allows for the user to buy as many tickets as they want.
        # For each ticket, winnings and tickets bought are adjusted accordingly using augmeneted assingment.
        winnings -= 2
        tickets_bought += 1
        
        # For each ticket, the list kept to store each tickets user selected lottery numbers will be emptied
        user_lotterynumbers = []
        
        
        print("")
        # When the user is selecting their lottery numbeers, the ticket number that they are on will be displayed
        print(f"You are selecting numbers for: Ticket Number {tickets_bought}\n")
    
    
        # User selection of lottery numbers for a given ticket
        while len(user_lotterynumbers) < 6:
            while len(user_lotterynumbers) < 5:
                # This nested while loop is designed to prompt the user to enter their 5 non-Powerball numbers
                try:
                    # Try propmts the user for an input and attempts to convert it to an integer value
                    user_number = int(input("Enter a non-powerball number that is between 1 and 69: "))
                except ValueError:
                    # except ValueError displays an invalid input message if the user attempts to enter a datatype other than an integer
                    print("Invalid input. Input must be a number that is bewteen 1 and 69.")
                else:
                    # If the user input is an integer value, the else statment evaluates 
                    # if the integer meets the proper criteria for the non-powerball numbers
                    if 1 <= user_number <= 69:
                        if user_number not in user_lotterynumbers:
                            user_lotterynumbers.append(user_number)
                        else:
                            print(f"Invalid input. Already entered {user_number} for this ticket.")
                            print(f"No duplicate numbers allowed. The second {user_number} will not be added to ticket.")
                    else:
                        print(f"Invalid input. {user_number} is not between 1 and 69.")
                        print(f"Non-Powerball number {user_number} will not be added to this ticket.")
                        
            user_lotterynumbers.sort()
            
            try:
                 # Try propmts the user for an input and attempts to convert it to an integer value
                user_powerball = int(input("Enter your powerball number that is between 1 and 26: "))
            except ValueError:
                # except ValueError displays an invalid input message if the user attempts to enter a datatype other than an integer
                print("Invalid input. Input must be a number that is between 1 and 69.")
            else:
                # If the user input is an integer value, the else statment evaluates 
                # if the integer meets the proper criteria for the powerball number
                if 1 <= user_powerball <= 26:
                    user_lotterynumbers.append(user_powerball)
                    User_tickets.update({tickets_bought: user_lotterynumbers})
                    # Once the user enters 5 proper non-powerball numbers and a powerball number, 
                    # the ticket number and list of user selected numbers is added to the dictionary
                else:
                    print(f"Invalid input. {user_powerball} is not between 1 and 26.")
                    print(f"Powerball number {user_powerball} will not be added to this ticket.")
        
        print("")
        
        # Prompts the user if they want to play again
        while True:
            decision_to_continue = input("Would you like to buy another ticket? (Yes or No): ")
            if decision_to_continue.lower() == "yes":
                print("You decided to buy another ticket. \n")
                break
            elif decision_to_continue.lower() == "no":
                print("You decided not to but another ticket. \n")
                break  
            else:
                print("Invalid input. Type Yes to continue playing. Type No to stop playing.")
                
        # If the user decides not to play, then the program generates the winning lottery numbers.
        if decision_to_continue.lower() == "no":
            winning_lotterynumbers = []
            
            while len(winning_lotterynumbers) < 6:
                while len(winning_lotterynumbers) < 5:
                    winning_number = random.randrange(1, 70)
                    if winning_number not in winning_lotterynumbers:
                        winning_lotterynumbers.append(winning_number)
                        
                winning_lotterynumbers.sort()
                winning_powerball = random.randrange(1, 27)
                winning_lotterynumbers.append(winning_powerball)
                

            # Printing of the tickets that the user had bought
            for ticket_number, user_number in User_tickets.items():
                print("-------------------------------------------------------")
                print(f"Ticket Number: {ticket_number}")
                
                # Ticket displays the user selected numbers
                print("Your selected numbers are: ")
                for number in user_number:
                    print(number, end=" ")
                
                print("")
                
                # Ticket displays the randomly generated winning lottery numbers
                print("The winning numbers are: ")
                for number in winning_lotterynumbers:
                    print(number, end=" ")
                    
                print("\n")
                
                print("Results:")
                
                # Matching evaluates the number of matching numbers between the user selected lottery 
                # and winning non-powerball numbers for each ticket
                matching = 0
                for number in user_number[:5]:
                    if number in winning_lotterynumbers[:5]:
                        matching += 1
                
                # If-elif statement for the possible number of matching non-powerball numbers
                # Nested if-else statements evaluates if the user powerball number and winning powerball number are matching
                # Print statement displays the type of matching combination and its corresponding winnings
                if matching == 0:
                    if user_number[5] == winning_lotterynumbers[5]:
                        print("You matched only the powerball number")
                        print("This result corresponds to $4.00 in winnings")
                        winnings += 4
                    else:
                        print("You did not have any matching numbers")
                        print("This result does not correspond to any winnings")
                    
                elif matching == 1:
                    if user_number[5] == winning_lotterynumbers[5]:
                        print("You matched 1 out of the 5 non-powerball numbers and the powerball number")
                        print("This result corresponds to $4.00 in winnings")
                        winnings += 4
                    else:
                        print("You only matched 1 out of the 5 non-powerball numbers")
                        print("This result does not correspond to any winnings")
                        
                elif matching == 2:
                    if user_number[5] == winning_lotterynumbers[5]:
                        print("You matched 2 out of the 5 non-powerball numers and the powerball number")
                        print("This result corresponds to $7.00 in winnings")
                        winnings += 7
                    else:
                        print("You only matched 2 out of the 5 non-powerball numbers")
                        print("This result does not correspond to any winnings")
                    
                elif matching == 3:
                    if user_number[5] == winning_lotterynumbers[5]:
                        print("You matched 3 out of the 5 non-powerball numbers and the powerball number")
                        print("This result corresponds to $100.00 in winnings")
                        winnings += 100
                    else:
                        print("You only matched 3 out of the 5 non-powerball numbers")
                        print("This result corresponds to $7.00 in winnings")
                        winnings += 7
                    
                elif matching == 4:
                    if user_number[5] == winning_lotterynumbers[5]:
                        print("You matched 4 out of the 5 non-powerball numbers and the powerball number")
                        print("This result corresponds to $50,000.00 in winnings")
                        winnings += 50000
                    else:
                        print("You only matched 4 out of the 5 non-powerball numbers")
                        print("This result corresponds to $100.00 in winnings")
                        winnings += 100
                    
                elif matching == 5:
                    if user_number[5] == winning_lotterynumbers[5]:
                        print("You matched all 5 non-powerball numbers and the powerball number")
                        print("You hit the jackpot!! Current jackpot: $100,000,000")
                        winnings += 100000000
                    else:
                        print("You only matched the 5 non-powerball numbers")
                        print("This result corresponds to $1,000,000 in winnings")
                        winnings += 1000000
                print("------------------------------------------------------- \n")
                
            # Once every ticket is displayed the overall number of tickets bought and winnings are displayed.
            if winnings > 0:
                print(f"Tickets bought: {tickets_bought}")
                print(f"Overall winnings: +${winnings}.00")
                
            elif winnings == 0:
                print(f"Tickets bought: {tickets_bought}")
                print(f"Overall winnings: ${winnings}.00")
                
            else:
                print(f"Tickets bought: {tickets_bought}")
                print(f"Overall winnings: -${abs(winnings)}.00")
                
            break


Powerball()
```
